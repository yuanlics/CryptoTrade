```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

PRICE_TIME_FMT = "%Y-%m-%d %H:%M:%S UTC"
starting_date = "2023-08-01"
ending_date = "2024-02-01"
PATH_PRICE = 'data/eth_daily.csv'
df = pd.read_csv(PATH_PRICE)
df['date'] = pd.to_datetime(df['snapped_at'], format=PRICE_TIME_FMT)

df = df[(df['date'] >= starting_date) & (df['date'] <= ending_date)]  # only use ending_date for open price

# visualize open price
plt.figure(figsize=(14, 7))
plt.plot(df['date'], df['open'])
plt.title('ETH Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
plt.show()
```
